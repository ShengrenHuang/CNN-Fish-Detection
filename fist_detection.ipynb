```python
from os import listdir
location_of_data = 'fish_dataset'
listdir(location_of_data)
```

```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
base_path = 'fish_dataset/'
sample_image = "Black Sea Sprat_1.png"
image = read_image(base_path+sample_image)
plt.imshow(image)
plt.show()
```

```python
from detecto.core import Dataset
dataset = Dataset(base_path)
```

```python
from detecto.visualize import show_labeled_image

image, targets = dataset[10]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
from detecto.core import DataLoader, Model

labels = ['BSS', 'GHB', 'HM'] # BSS = Black Sea Sprat; GHB = Gilt-Head Bream; HM = Hourse Mackerel

model = Model(labels)
```

```python
model.fit(dataset, learning_rate = 0.0001, epochs = 20 )
```

```python
test_image = '00027.png'
image = read_image(test_image)
label, boxes, scores = model.predict(image)
```

```python
print('labels', labels)
print('boxes', boxes)
print('scores', scores)
```
